```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10, 105, 10)
y = x**2

print(list(x))
print(list(y))
plt.title("Sample Graph")
plt.xlabel("Distance (km)")
plt.ylabel("Time (m/s)")
plt.plot(x,y,"y")
plt.plot(x, y,"o")
plt.grid()
```

```python
x = np.arange(0,10)
y = x ^ 2

plt.subplot(1,2,1)
plt.plot(x,y)

x = np.arange(10, 20)
y = x ^ 2

plt.subplot(1,2,2)
plt.plot(x,y)
```

```python
fruits = ["orange", "banana", "apple", "mango", "pineapple"]
quantity = [15,10,25,20,30]
explode = (0,0,0,0,0.1)
#ax1 = plt.subplots()
plt.pie(quantity, labels=fruits, explode=explode, autopct="%1.1f%%", shadow=True, startangle=90)
plt.axis("equal")
plt.show()
```

```python
fruits = ["orange", "banana", "apple", "mango", "pineapple"]
quantity = [15,10,25,20,30]
plt.xlabel("fruits")
plt.ylabel("Quantity")
plt.title("Fruit market stats")
plt.bar(fruits,quantity)
plt.show()
```

```python
students = ["male", "female"]
quantity = [15,10]
explode = (0,0,)
#ax1 = plt.subplots()
plt.pie(quantity, labels=students, explode=explode, autopct="%1.1f%%", shadow=True, startangle=90)
plt.axis("equal")
plt.title("Distribution of male and females in a class")
plt.show()
```
