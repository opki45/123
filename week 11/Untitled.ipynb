```python
df = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/IT_Salary_Survey_EU_18-20/Survey_2019.csv")
df = df.fillna(method="ffill")
#print(df.isna().any())
#print(df.head())
plt.figure(figsize=(16, 12))
sns.heatmap(df.corr(), cmap="bwr", annot=True)
#print(df["City"].unique())
df = df.loc[(df["City"].str.contains("Berlin")) | (df["City"].str.contains("Munich"))]

df = df.fillna(method="ffill")
print(df[["Age", "Years of experience", "Yearly bonus", "Yearly brutto salary (without bonus and stocks)","Number of vacation days" ]])
#print(df.columns)

y = df["Yearly brutto salary (without bonus and stocks)"]
x = df[["Age", "Years of experience", "Yearly bonus", "Number of vacation days"]]

regression = linear_model.LinearRegression()
regression.fit(x,y)

predicted_salary = regression.predict([[33, 13, 1000, 29]])
print(predicted_salary)
print(regression.coef_)
```
